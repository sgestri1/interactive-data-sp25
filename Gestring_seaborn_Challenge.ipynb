## Main Assignment

First I installed the necessary modules for this challenge.

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

Then I loaded in the necessary csv file into a pandas DataFrame.

```python
ins = pd.read_csv('insurance.csv')
ins.head()
```

```python
ins.shape
```

```python
ins.dtypes
```

### Scatterplots

Here I created a scatterplot using the relplot function, with ‘bmi’ as the x-axis, ‘charges’ as the y-axis, and ‘sex’ as the column.

```python
sns.relplot(data=ins, x="bmi", y="charges", col="sex", kind="scatter")
```

Then I created nearly the same scatterplot but this time with the "region" variable set as the size argument for the graph.

```python
sns.relplot(data=ins, x="bmi", y="charges", col="sex", kind="scatter", size="region")
```

### Map Function

Here I am using the map function to create a new column in the DataFrame for smoker type, with two values: Smoker or Non-smoker. This will allow us to compare Smokers and Non-smokers using those labels rather than yes/no which may be less clear.

```python
ins["smoker_ind"]=ins["smoker"].map({"yes":"Smoker","no":"Non-smoker"})
```

```python
ins.head()
```

### Final Scatterplot

Now that I have my Smoker/Non-smoker columm, I can represent that data as colors in a new scatterplot using the "hue" argument.

```python
sns.relplot(data=ins, x="bmi", y="charges", col="sex", hue="smoker_ind", kind="scatter", size="region", alpha=0.5)
```
